```python
class Perceptron(object):
    
    def __init__(self, learning_rate=0.01, epochs=50):
        
        self.learning_rate = learning_rate
        self.epochs = epochs
        
    
    def _sum(self, X):
       
        dp = np.dot(X, self._weights) + self._bias
        
        return dp
    
    def _activation(self, value):
        return np.where(value >= 0.0, 1, -1)
    
    def fit(self, X, y):
        
        self._bias = np.random.uniform(-1, 1)
        self._weights = np.random.uniform(-1, 1, (X.shape[1]))
        
        for epoch_number in range(self.epochs):
            errors = 0
                
            for xi, target in zip(X, y):
                
                output = self.predict(xi)
                update = self.learning_rate * (target - output)
                    
                self._bias += update
                
                self._weights += update * xi
                
                errors += int(update != 0.0)
                          
        return self
    
    def getWeights(self):
        return self._weights
    
    def predict(self, X):
        result = self._activation(self._sum(X))
        
        return result
```

```python
a, b = 2.358, 0.15
```

```python
x = np.random.uniform(-1,1,100)
y = np.random.uniform(-1,1,100)

uvas = y < a*x+b
bananas = y >= a*x+b

labels = np.array([-1]*100)
labels[bananas] = 1
```

```python
X = np.array([[x[i], y[i]] for i in np.arange(100)])
```

```python
plt.plot(x[uvas],y[uvas],'o', label = 'Uvas')
plt.plot(x[bananas],y[bananas],'o', label = 'Bananas')

plt.xlabel('Peso')
plt.ylabel('Tamanho')

plt.legend()

plt.savefig(f"data-frutas", dpi=150)
```

```python
ppn = Perceptron(0.01, 100)
ppn.fit(X, labels)
```

```python
ppn.getWeights()
xl = np.linspace(-1,1,  50)
```

```python
plt.plot(x[uvas],y[uvas],'o', label = 'Uvas')
plt.plot(x[bananas],y[bananas],'o', label = 'Bananas')

plt.plot(xl, a*xl+b, label = f'{a}x + {b}= y')

plt.legend()

plt.savefig(f"frutas-result", dpi=150)
```
